# **3. Visualisasi Data**

Visualisasi Histogram

```python
# Visualisasi data Histogram
# Menghitung jumlah baris dan kolom untuk tata letak subplot
num_cols = 3
num_rows = (len(diamonds_df.columns) - 1) // num_cols + 1

# Membuat subplot
fig, axes = plt.subplots(num_rows, num_cols, figsize=(15, 10))

# Meratakan sumbu untuk memudahkan iterasi
axes = axes.flatten()

# Plot histogram untuk setiap kolom
for i, column in enumerate(diamonds_df.columns):
    if column != 'Unnamed: 0':  # Exclude non-numeric columns
        ax = axes[i]
        ax.hist(diamonds_df[column], bins=30, color='cyan', edgecolor='black')
        ax.set_title(column)
        ax.set_xlabel(column)
        ax.set_ylabel('Frequency')

# Sembunyikan subplot kosong
for i in range(len(diamonds_df.columns), num_cols * num_rows):
    fig.delaxes(axes[i])

# Menyesuaikan tata letak
plt.tight_layout()

# Menampilkan plots
plt.show()
```

Visualisasi Correlation Matrix(Heatmap)

```python
# Menghitung korelasi antar fitur
correlation_matrix = diamonds_df.corr()

# Plotting heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='GnBu', fmt=".2f", linewidths=0.5)
plt.title('Heatmap of Correlation Matrix')
plt.show()
```

Visualisasi Boxplot

```python
# Visualisasi box plot untuk setiap kolom
num_cols = 3
num_rows = (len(diamonds_df.columns) - 1) // num_cols + 1

plt.figure(figsize=(15, 10))
for i, column in enumerate(diamonds_df.columns):
    if column != 'Unnamed: 0' and diamonds_df[column].dtype in ['int64', 'float64']:  # Exclude non-numeric columns
        plt.subplot(num_rows, num_cols, i + 1)
        plt.boxplot(diamonds_df[column], vert=False)  # Mengatur vert=False untuk boxplot horizontal
        plt.title(column)

plt.tight_layout()
plt.show()
```

Visualisasi Scatter

```python
# Visualisasi scatter plot untuk setiap kolom numerik
num_cols = 3
num_rows = (len(diamonds_df.columns) - 1) // num_cols + 1

plt.figure(figsize=(15, 10))
for i, column in enumerate(diamonds_df.columns):
    if column != 'Unnamed: 0' and diamonds_df[column].dtype in ['int64', 'float64']:  # Exclude non-numeric columns
        plt.subplot(num_rows, num_cols, i + 1)
        plt.scatter(diamonds_df[column], diamonds_df['price'], alpha=0.5)
        plt.title(column)
        plt.xlabel(column)
        plt.ylabel('Price')
plt.tight_layout()
plt.show()
```

Visualisasi Violin Plot

```python
num_cols = 3
num_rows = (len(diamonds_df.columns) - 1) // num_cols + 1

plt.figure(figsize=(15, 10))
for i, column in enumerate(diamonds_df.columns):
    if column != 'Unnamed: 0' and diamonds_df[column].dtype in ['int64', 'float64']:  # Exclude non-numeric columns
        plt.subplot(num_rows, num_cols, i + 1)
        sns.violinplot(data=diamonds_df, y=column, palette='viridis')
        plt.title(f'Violin Plot of {column}')
        plt.ylabel(column)
plt.tight_layout()
plt.show()
```
