# **3. Visualisasi Data**

Visualisasi Histogram

```python
# Visualisasi data Histogram
first_10_columns = labs_df.iloc[:, :10]

plt.figure(figsize=(15, 10))
for i, column in enumerate(first_10_columns.columns, 1):
    plt.subplot(3, 4, i)
    plt.hist(first_10_columns[column], bins=20, color='skyblue', edgecolor='black')
    plt.title(column)
    plt.xlabel('Value')
    plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

Visualisasi Box Plot

```python
# Select columns suitable for boxplot
suitable_columns = labs_df[['SEQN', 'URXCRS', 'LBXAPB', 'LBXSSE', 'PHACOFHR', 'PHACOFMN', 'PHAANTHR', 'URXVOL2']]

# Membuat box plot untuk setiap fitur numerik
fig, axes = plt.subplots(2, 4, figsize=(15, 10))
axes = axes.flatten()

for i, feature in enumerate(suitable_columns):
    sns.boxplot(x=labs_df[feature], ax=axes[i])
    axes[i].set_title(f'{feature}')
    axes[i].set_xlabel(feature)

plt.tight_layout()
plt.show()
```


Visualisasi Correlation Matrix(Heatmap)

```python
# Select columns suitable for correlation heatmap
suitable_columns = labs_df[['SEQN', 'URXCRS', 'LBXAPB', 'LBXSSE', 'PHACOFHR', 'PHACOFMN', 'PHAANTHR', 'URXVOL2']]

# Calculate correlation matrix
correlation_matrix = suitable_columns.corr()

# Plot heatmap of correlation matrix
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='GnBu', fmt=".2f")
plt.title('Heatmap Correlation Matrix')
plt.show()
```

Visualisasi Scatter

```python
# Select columns suitable for scatter plot
suitable_columns = labs_df[['SEQN', 'URXCRS', 'LBXAPB', 'LBXSSE', 'PHACOFHR', 'PHACOFMN', 'PHAANTHR', 'URXVOL2']]

# Create pairplot
sns.pairplot(suitable_columns, hue='SEQN', palette='husl')
plt.suptitle('Pairplot of Suitable Columns', y=1.02)
plt.show()
```

Visualisasi Violin Plot

```python
# Select columns suitable for scatter plot
suitable_columns = labs_df[['SEQN', 'URXCRS', 'LBXAPB', 'LBXSSE', 'PHACOFHR', 'PHACOFMN', 'PHAANTHR', 'URXVOL2']]

# Create violin plot for each column
plt.figure(figsize=(8, 12))
for i, column in enumerate(suitable_columns.columns, 1):
    plt.subplot(4, 2, i)
    sns.violinplot(data=suitable_columns[column], inner="points", palette="muted")
    plt.title(f'Violin Plot of {column}')
    plt.xlabel(column)

plt.tight_layout()
plt.show()
```
