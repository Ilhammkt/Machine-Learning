```python
#1 Impor library yang diperlukan
import pandas as pd
from sklearn.cluster import AgglomerativeClustering
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#2 Membaca dataset dari file CSV
file_path = "/content/drive/MyDrive/Dataset/Mall_Customers.csv"
data = pd.read_csv(file_path)
```

```python
#3 Mengambil fitur yang akan digunakan untuk clustering
X = data.iloc[:, [3, 4]].values  # Menggunakan 'Annual Income' dan 'Spending Score'
```

```python
#4 Hierarchical clustering
hc = AgglomerativeClustering(n_clusters=5, affinity='euclidean', linkage='ward')
y_hc = hc.fit_predict(X)
```

```python
#5 Visualisasi
plt.figure(figsize=(10, 6))
sns.scatterplot(x=X[y_hc == 0, 0], y=X[y_hc == 0, 1], color='red', label='Cluster 1', s=50)
sns.scatterplot(x=X[y_hc == 1, 0], y=X[y_hc == 1, 1], color='blue', label='Cluster 2', s=50)
sns.scatterplot(x=X[y_hc == 2, 0], y=X[y_hc == 2, 1], color='green', label='Cluster 3', s=50)
sns.scatterplot(x=X[y_hc == 3, 0], y=X[y_hc == 3, 1], color='orange', label='Cluster 4', s=50)
sns.scatterplot(x=X[y_hc == 4, 0], y=X[y_hc == 4, 1], color='purple', label='Cluster 5', s=50)
plt.title('Clusters of customers')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.legend(loc='upper right')
plt.grid(True)
plt.show()
```
