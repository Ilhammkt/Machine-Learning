# Visualisasi Data


```python
# Visualisasi distribusi gaji
plt.figure(figsize=(10, 6))
sns.histplot(data['Yearly brutto salary (without bonus and stocks) in EUR'], bins=30, kde=True, color='skyblue')
plt.title('Distribusi Gaji')
plt.xlabel('Gaji (EUR)')
plt.ylabel('Frekuensi')
plt.show()
```

```python
# Visualisasi korelasi antara variabel numerik
numeric_data = data.select_dtypes(include=['int64', 'float64'])
plt.figure(figsize=(10, 6))
sns.heatmap(numeric_data.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Korelasi Antara Variabel Numerik')
plt.show()
```

```python
# Visualisasi hubungan antara gaji dan pengalaman kerja
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Total years of experience', y='Yearly brutto salary (without bonus and stocks) in EUR', data=data, color='green')
plt.title('Hubungan antara Gaji dan Pengalaman Kerja')
plt.xlabel('Total Tahun Pengalaman')
plt.ylabel('Gaji (EUR)')
plt.show()
```

```python
# Visualisasi boxplot gaji berdasarkan jenis kelamin
plt.figure(figsize=(10, 6))
sns.boxplot(x='Gender', y='Yearly brutto salary (without bonus and stocks) in EUR', data=data, palette='husl')
plt.title('Perbandingan Gaji Berdasarkan Jenis Kelamin')
plt.xlabel('Jenis Kelamin')
plt.ylabel('Gaji (EUR)')
plt.show()
```
